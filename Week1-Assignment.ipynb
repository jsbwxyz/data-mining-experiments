10. Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

Knowing that it'll return an error we premptively replace non-numeric strings with NaNs.

```python
video_games_df['User_Score'] = pd.to_numeric(video_games_df['User_Score'], errors='coerce')
```

The above ststment uses the `to_numeric()` function to convert all non-numeric rows in this column to NaN. This function is used to simply select all rows that are numeric.

The part `errors=coerce` is what actually changes the non-numeric values by forcing (coercing) any value that can't be converted to a number into NaN instead of throwing an error

```python
video_games_df['User_Score']
```

In the below statment, NaN values are found using the `fillna()` function and populated with the median of all scores by using `median()`.

```python
video_games_df['User_Score'] = video_games_df['User_Score'].fillna(video_games_df['User_Score'].median())
```
